```python
import requests

url = "https://leetcode.com/graphql/"
query = """
query problemsetQuestionList($categorySlug: String, $limit: Int, $skip: Int, $filters: QuestionListFilterInput) {
  problemsetQuestionList: questionList(
    categorySlug: $categorySlug
    limit: $limit
    skip: $skip
    filters: $filters
  ) {
    total: totalNum
    questions: data {
      acRate
      difficulty
      freqBar
      frontendQuestionId: questionFrontendId
      isFavor
      paidOnly: isPaidOnly
      status
      title
      titleSlug
      topicTags {
        name
        id
        slug
      }
      hasSolution
      hasVideoSolution
    }
  }
}
"""

headers = {
    "Content-Type": "application/json",
    "Authorization": "Bearer YOUR_ACCESS_TOKEN"  # Replace with your token if necessary
}

def fetch_questions(category_slug, limit, filters):
    skip = 0
    all_questions = []
    while True:
        variables = {
            "categorySlug": category_slug,
            "limit": limit,
            "skip": skip,
            "filters": filters
        }
        response = requests.post(url, json={'query': query, 'variables': variables}, headers=headers)
        data = response.json()
        questions = data.get('data', {}).get('problemsetQuestionList', {}).get('questions', [])
        if not questions:
            break
        all_questions.extend(questions)
        skip += limit
    return all_questions

category_slug = "all-code-essentials"
limit = 100
filters = {}

questions = fetch_questions(category_slug, limit, filters)
for question in questions:
    print(question)
```

```python
url = "https://leetcode.com/graphql/"
query = """
query questionContent($titleSlug: String!) {
  question(titleSlug: $titleSlug) {
    content
    mysqlSchemas
    dataSchemas
  }
}
"""

headers = {
    "Content-Type": "application/json",
    "Authorization": "Bearer YOUR_ACCESS_TOKEN"  # Replace with your token if necessary
}

def fetch_question_content(title_slug):
    variables = {
        "titleSlug": title_slug
    }
    response = requests.post(url, json={'query': query, 'variables': variables}, headers=headers)
    data = response.json()
    question_content = data.get('data', {}).get('question', {}).get('content', '')
    return question_content

# Initialize a new DataFrame to store the results
results_dict = {'titleSlug': [], 'question_content': []}

# Iterate through each row in the original DataFrame
for index, row in df.iterrows():
    title_slug = row['titleSlug']
    print(f"Fetching question content for {title_slug}...")
    question_content = fetch_question_content(title_slug)
    results_dict['titleSlug'].append(title_slug)
    results_dict['question_content'].append(question_content)


```

```python
df_questions = pd.DataFrame(results_dict)
df_questions
```

```python
df_questions.to_json('all_leetcode_questions_text.json', orient='records')
```

```python
import json

# Function to transform the list of dictionaries to a single dictionary
def transform_json(input_file, output_file):
    # Read the JSON data from the file
    with open(input_file, 'r') as f:
        data = json.load(f)
    
    # Transform the list of dictionaries to a single dictionary
    transformed_data = {item['titleSlug']: item['question_content'] for item in data}
    
    # Write the transformed data to a new JSON file
    with open(output_file, 'w') as f:
        json.dump(transformed_data, f, indent=4)

# Specify the input and output file paths
input_file = output_file = 'all_leetcode_questions_text.json'

# Call the function to transform and save the JSON data
transform_json(input_file, output_file)

print(f'Transformed data has been written to {output_file}')
```
